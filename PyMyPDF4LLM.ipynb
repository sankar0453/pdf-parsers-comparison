## Extract specific pages

```python
md_text_pages = pymupdf4llm.to_markdown("/content/asco-4556.pdf", pages=[0])
```

```python
md_text_pages
```

```python
md_text_chunks = pymupdf4llm.to_markdown(
    "/content/asco-4556.pdf",
    pages=[0],
    page_chunks=True   )
md_text_chunks
```

```python
md_text_words = pymupdf4llm.to_markdown(
    "/content/asco-4556.pdf",
    pages=[0],
    # chunk_size=1000,
    page_chunks=True,
    # write_images=True,
    extract_words=True,
    # chunk_overlap=50
    )
md_text_words
```

```python
md_text_tables = pymupdf4llm.to_markdown(
    "/content/asco-4556.pdf",
    # page_chunks=True,
    extract_tables=True
    )
md_text_tables
```
